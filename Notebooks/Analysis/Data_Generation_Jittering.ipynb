```python
import pandas as pd
```

```python
# 20 Newsgroups
df_20newsgroup = pd.read_csv("Stability_Jittering/results_experiment_4_20_newsgroups.csv")
df_high_20newsgroup = pd.read_csv("Stability_Jittering/results_experiment_5_corpus_similarity_20_newsgroups.csv")

df_20newsgroup["alpha"] = (0.25*(df_20newsgroup["trustworthiness"] + df_20newsgroup["continuity"] + df_20newsgroup["mrre_missing"] + df_20newsgroup["mrre_false"]) + df_20newsgroup['label_preservation'] + df_20newsgroup["local_continuity"])/3.0
df_20newsgroup["beta"] = 0.5 * (0.5 * (0.5*(df_20newsgroup["pearson_correlation"]+1) + 0.5*(df_20newsgroup["spearman_correlation"]+1)) + 0.5*(df_20newsgroup["cluster_ordering"]+1))
df_20newsgroup["gamma"] = 1 - (df_20newsgroup["distance_consistency"]) 


df_high_20newsgroup["alpha"] = (0.25*(df_high_20newsgroup["trustworthiness"] + df_high_20newsgroup["continuity"] + df_high_20newsgroup["mrre_missing"] + df_high_20newsgroup["mrre_false"]) + df_high_20newsgroup['label_preservation'] + df_high_20newsgroup["local_continuity"])/3.0
df_high_20newsgroup["beta"] = 0.5 * (0.5 * (0.5*(df_high_20newsgroup["pearson_correlation"]+1) + 0.5*(df_high_20newsgroup["spearman_correlation"]+1)) + 0.5*(df_high_20newsgroup["cluster_ordering"]+1))
df_high_20newsgroup["gamma"] = 1 - (df_high_20newsgroup["distance_consistency"]) 

n,m = df_20newsgroup.shape
list_jitter1 = []
list_jitter2 = []

a_list = df_20newsgroup["jitter_1"].tolist()
b_list = df_20newsgroup["jitter_2"].tolist()

for i in range(n):
    a = a_list[i]
    b = b_list[i]
    if a > b:
        list_jitter1.append(b)
        list_jitter2.append(a)
    elif b > a:
        list_jitter1.append(a)
        list_jitter2.append(b)
    else:
        print("Error")


df_20newsgroup["jitter1_correct"] = list_jitter1
df_20newsgroup["jitter2_correct"] = list_jitter2

alpha_adjust = []
beta_adjust = []
gamma_adjust = []

alpha_jitter010 = df_high_20newsgroup['alpha'].tolist()[0]
alpha_jitter025 = df_high_20newsgroup['alpha'].tolist()[1]
beta_jitter010 = df_high_20newsgroup['beta'].tolist()[0]
beta_jitter025 = df_high_20newsgroup['beta'].tolist()[1]
gamma_jitter010 = df_high_20newsgroup['gamma'].tolist()[0]
gamma_jitter025 = df_high_20newsgroup['gamma'].tolist()[1]

for i in range(n):
    if list_jitter2[i] == 0.1:
        alpha_adjust.append(1 - abs(df_20newsgroup['alpha'].tolist()[i] - alpha_jitter010))
        beta_adjust.append(1 - abs(df_20newsgroup['beta'].tolist()[i] - beta_jitter010))
        gamma_adjust.append(1 - abs(df_20newsgroup['gamma'].tolist()[i] - gamma_jitter010))
    elif list_jitter2[i] == 0.25:
        alpha_adjust.append(1 - abs(df_20newsgroup['alpha'].tolist()[i] - alpha_jitter025))
        beta_adjust.append(1 - abs(df_20newsgroup['beta'].tolist()[i] - beta_jitter025))
        gamma_adjust.append(1 - abs(df_20newsgroup['gamma'].tolist()[i] - gamma_jitter025))
    else:
        print("Error")


df_20newsgroup['TM'].replace({
    'bert_all_distilroberta_v1': 'BERT',
    'bert_all_mpnet_base_v2': 'BERT',
    'bow': 'BOW',
    'tfidf': 'BOW',
    'lda': 'LDA',
    'lda_linear_combined': 'LDA',
    'lsi': 'LSI',
    'lsi_linear_combined': 'LSI',
    'lsi_tfidf': 'LSI',
    'lsi_linear_combined_tfidf': 'LSI',
    'nmf': 'NMF',
    'nmf_tfidf': 'NMF',
    'nmf_linear_combined': 'NMF',
    'nmf_linear_combined_tfidf': 'NMF',
    'doc2vec': 'Doc2Vec'
}, inplace=True)

df_20newsgroup["alpha_adjust"] = alpha_adjust
df_20newsgroup["beta_adjust"] = beta_adjust
df_20newsgroup["gamma_adjust"] = gamma_adjust
```
